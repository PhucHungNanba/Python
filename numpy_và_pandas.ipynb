#### PANDAS
###### 1. Tạo DataFrame từ Dictionary

```python
import pandas as pd

# Dữ liệu dạng dictionary
data = {
    'Ten': ['Hưng', 'Hội', 'Đăng', 'Quyền'],
    'Tuoi': [21, 16, 20, 22],
    'ThanhPho': ['Vĩnh Long', 'Bình Định', 'Tiền Giang', 'Ngã Ba Chuồng Chó']
}

# Tạo DataFrame
df = pd.DataFrame(data)

print(df)
```

##### 2. Đọc dữ liệu từ file CSV

```python
import pandas as pd

# Ví dụ đọc từ một URL
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df_titanic = pd.read_csv(url)

# Hiển thị dòng đầu tiên để xem qua dữ liệu
print(df_titanic.head())
```

##### 3. Xem thông tin tổng quan của DataFrame

```python
# Sử dụng DataFrame từ ví dụ 2
df_titanic.info()
```

##### 4. Lọc dữ liệu theo điều kiện

```python
# Sử dụng DataFrame từ ví dụ 1
# Lọc ra những người lớn hơn 23 tuổi
df_lon_hon_23 = df[df['Tuoi'] > 23]
print("Người lớn hơn 23 tuổi:")
print(df_lon_hon_23)

# Lọc những người ở Hà Nội VÀ lớn hơn 25 tuổi
df_loc_phuc_tap = df[(df['ThanhPho'] == 'Hà Nội') & (df['Tuoi'] > 25)]
print("\nNgười ở Hà Nội và lớn hơn 25 tuổi:")
print(df_loc_phuc_tap)
```

##### 5. Thêm một cột mới

```python
# Sử dụng DataFrame từ ví dụ 1
# Thêm cột 'NamSinh' được tính từ cột 'Tuoi'
nam_hien_tai = 2025
df['NamSinh'] = nam_hien_tai - df['Tuoi']

print(df)
```

6. Xử lý dữ liệu bị thiếu (Missing Data)

```python
# Sử dụng df_titanic từ ví dụ 2, cột 'Age' có nhiều giá trị thiếu
# Đếm số giá trị thiếu trong mỗi cột
print("Số giá trị thiếu ban đầu:")
print(df_titanic.isnull().sum())

# Cách 1: Xóa các hàng có giá trị thiếu ở cột 'Age'
# df_titanic.dropna(subset=['Age'], inplace=True)

# Cách 2: Điền giá trị trung bình vào các ô bị thiếu của cột 'Age'
gia_tri_trung_binh_tuoi = df_titanic['Age'].mean()
df_titanic['Age'].fillna(gia_tri_trung_binh_tuoi, inplace=True)

print("\nSố giá trị thiếu sau khi điền:")
print(df_titanic.isnull().sum())
```

7. Nhóm dữ liệu với groupby()

```python
# Sử dụng df_titanic
# Tính tuổi trung bình của hành khách theo giới tính ('Sex')
tuoi_tb_theo_gioi_tinh = df_titanic.groupby('Sex')['Age'].mean()

print(tuoi_tb_theo_gioi_tinh)
```

8. Sắp xếp dữ liệu

```python
# Sử dụng DataFrame từ ví dụ 1
# Sắp xếp theo tuổi giảm dần
df_sap_xep = df.sort_values(by='Tuoi', ascending=False)

print(df_sap_xep)
```

9. Hợp nhất (Merge) hai DataFrame

```python
# Tạo 2 DataFrame mẫu
df_info = pd.DataFrame({
    'ID': [1, 2, 3],
    'Ten': ['An', 'Binh', 'Chi']
})

df_diem = pd.DataFrame({
    'ID': [1, 2, 3],
    'DiemToan': [9, 7, 8],
    'DiemVan': [8, 9, 7]
})

# Hợp nhất 2 DataFrame dựa trên cột 'ID'
df_merged = pd.merge(df_info, df_diem, on='ID')

print(df_merged)
```

10. Áp dụng hàm tùy chỉnh với apply()

```python
# Sử dụng DataFrame từ ví dụ 1
# Tạo một hàm để phân loại độ tuổi
def phan_loai_tuoi(tuoi):
    if tuoi < 22:
        return 'Trẻ'
    elif 22 <= tuoi <= 28:
        return 'Thanh niên'
    else:
        return 'Trưởng thành'

# Áp dụng hàm này lên cột 'Tuoi' để tạo cột mới 'NhomTuoi'
df['NhomTuoi'] = df['Tuoi'].apply(phan_loai_tuoi)

print(df)
```
