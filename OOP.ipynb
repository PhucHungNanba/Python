```python
from abc import ABC, abstractmethod

# --- LỚP CHA: SINHVIEN ---
class SinhVien(ABC):
    """
    Lớp trừu tượng (Abstract Class) đại diện cho một sinh viên.
    """
    def __init__(self, ma_sv, ho_ten, ngay_sinh, diem_ren_luyen, diem_GPA):
        self._ma_sv = ma_sv
        self._ho_ten = ho_ten
        self._ngay_sinh = ngay_sinh
        self._diem_ren_luyen = diem_ren_luyen
        self._diem_GPA = diem_GPA

    # --- Getter/Setter (Properties) ---
    @property
    def ma_sv(self):
        return self._ma_sv

    @property
    def ho_ten(self):
        return self._ho_ten

    @ho_ten.setter
    def ho_ten(self, ten_moi):
        self._ho_ten = ten_moi

    @property
    def ngay_sinh(self):
        return self._ngay_sinh

    @ngay_sinh.setter
    def ngay_sinh(self, ngay_moi):
        self._ngay_sinh = ngay_moi

    @property
    def diem_ren_luyen(self):
        return self._diem_ren_luyen

    @diem_ren_luyen.setter
    def diem_ren_luyen(self, diem_rl_moi):
        self._diem_ren_luyen = diem_rl_moi

    @property
    def diem_GPA(self):
        return self._diem_GPA

    @diem_GPA.setter
    def diem_GPA(self, GPA_moi):
        self._diem_GPA = GPA_moi

    # --- Phương thức trừu tượng ---
    @abstractmethod
    def hien_thi_thong_tin(self):
        """Hiển thị thông tin chi tiết của sinh viên."""
        # Thêm phần hiển thị chung ở đây để lớp con kế thừa
        print(f"  Mã SV: {self._ma_sv}")
        print(f"  Họ Tên: {self._ho_ten}")
        print(f"  Ngày Sinh: {self._ngay_sinh}")
        print(f"  Điểm Rèn Luyện: {self._diem_ren_luyen}")
        print(f"  Điểm GPA: {self._diem_GPA}")

    # --- ĐÃ XÓA tinh_diem_thuong ---


# --- LỚP CON 1: SINH VIÊN ĐẠI TRÀ ---
class SinhVienDaiTra(SinhVien):
    """Lớp đại diện cho sinh viên hệ đại trà."""
    def __init__(self, ma_sv, ho_ten, ngay_sinh, diem_ren_luyen, diem_GPA):
        super().__init__(ma_sv, ho_ten, ngay_sinh, diem_ren_luyen, diem_GPA)

    def hien_thi_thong_tin(self):
        print(f"--- Sinh Viên Đại Trà ---")
        super().hien_thi_thong_tin() # Gọi phần hiển thị chung từ lớp cha

    # --- ĐÃ XÓA tinh_diem_thuong ---


# --- LỚP CON 2: SINH VIÊN CHẤT LƯỢNG CAO ---
class SinhVienChatLuongCao(SinhVien):
    """Lớp đại diện cho sinh viên hệ chất lượng cao."""
    def __init__(self, ma_sv, ho_ten, ngay_sinh, diem_ren_luyen, diem_GPA, diem_tieng_anh):
        super().__init__(ma_sv, ho_ten, ngay_sinh, diem_ren_luyen, diem_GPA)
        self.diem_tieng_anh = diem_tieng_anh

    def hien_thi_thong_tin(self):
        print(f"--- Sinh Viên Chất Lượng Cao ---")
        super().hien_thi_thong_tin() # Gọi phần hiển thị chung từ lớp cha
        print(f"  Điểm Tiếng Anh: {self.diem_tieng_anh}") # Chỉ in thêm phần riêng

    # --- ĐÃ XÓA tinh_diem_thuong ---


# --- LỚP QUẢN LÝ SINH VIÊN ---
class QuanLySinhVien:
    def __init__(self):
        self.danh_sach_sv = []

    def tim_sinh_vien_theo_ma(self, ma_sv):
        """Hàm hỗ trợ: Tìm và trả về đối tượng sinh viên nếu có."""
        for sv in self.danh_sach_sv:
            if sv.ma_sv == ma_sv:
                return sv
        return None

    def them_sinh_vien(self):
        """Chức năng 1: Thêm sinh viên mới."""
        print("\n--- Thêm Sinh Viên Mới ---")
        ma_sv = input("Nhập mã SV: ").strip()

        if self.tim_sinh_vien_theo_ma(ma_sv):
            print("LỖI: Mã sinh viên này đã tồn tại!")
            return

        ho_ten = input("Nhập họ tên: ").strip()
        ngay_sinh = input("Nhập ngày sinh (dd/mm/yyyy): ").strip()

        try:
            diem_ren_luyen = float(input("Nhập điểm rèn luyện: "))
            diem_GPA = float(input("Nhập điểm GPA (hệ 4): "))
            if not (0.0 <= diem_GPA <= 4.0):
                print("Lỗi: Điểm GPA phải từ 0.0 đến 4.0.")
                return
        except ValueError:
            print("Lỗi: Điểm phải là một con số.")
            return

        loai_sv = input("Loại sinh viên (1: Đại trà, 2: Chất lượng cao): ").strip()

        if loai_sv == '1' :
            sv_moi = SinhVienDaiTra(ma_sv, ho_ten, ngay_sinh, diem_ren_luyen, diem_GPA)

        elif loai_sv == '2':
            try:
                diem_tieng_anh = float(input("Nhập điểm Tiếng Anh: "))
            except ValueError:
                print("Lỗi: Điểm phải là một con số.")
                return

            sv_moi = SinhVienChatLuongCao(ma_sv, ho_ten, ngay_sinh, diem_ren_luyen, diem_GPA, diem_tieng_anh)

        else:
            print("Lựa chọn không hợp lệ.")
            return

        self.danh_sach_sv.append(sv_moi)
        print(f"Đã thêm thành công sinh viên {ho_ten} (Mã: {ma_sv}).")

    def xoa_sinh_vien(self):
        """Chức năng 2: Xóa sinh viên."""
        print("\n--- Xóa Sinh Viên ---")
        ma_sv = input("Nhập mã SV cần xóa: ").strip()

        sv = self.tim_sinh_vien_theo_ma(ma_sv)

        if sv:
            confirm = input(f"Bạn có chắc muốn xóa sinh viên {sv.ho_ten} (Mã: {ma_sv})? (Có/Không): ").lower()
            if confirm == 'có' or confirm == 'co':
                self.danh_sach_sv.remove(sv)
                print("Đã xóa sinh viên thành công.")
            else:
                print("Đã hủy thao tác xóa.")
        else:
            print("LỖI: Không tìm thấy sinh viên có mã này.")

    def sua_thong_tin_sinh_vien(self):
        """Chức năng 3: Sửa thông tin sinh viên."""
        print("\n--- Sửa Thông Tin Sinh Viên ---")
        ma_sv = input("Nhập mã SV cần sửa: ").strip()

        sv = self.tim_sinh_vien_theo_ma(ma_sv)

        if sv:
            print(f"Đang sửa thông tin cho: {sv.ho_ten}")
            print("(Bỏ trống và nhấn Enter nếu không muốn thay đổi)")

            try:
                # Sửa thông tin chung
                ho_ten_moi = input(f"Họ tên mới ({sv.ho_ten}): ").strip()
                if ho_ten_moi:
                    sv.ho_ten = ho_ten_moi

                ngay_sinh_moi = input(f"Ngày sinh mới ({sv.ngay_sinh}): ").strip()
                if ngay_sinh_moi:
                    sv.ngay_sinh = ngay_sinh_moi

                diem_ren_luyen_moi = input(f"Điểm rèn luyện mới ({sv.diem_ren_luyen}): ").strip()
                if diem_ren_luyen_moi:
                    sv.diem_ren_luyen = float(diem_ren_luyen_moi)

                diem_GPA_moi = input(f"Điểm GPA mới ({sv.diem_GPA}): ").strip()
                if diem_GPA_moi:
                    sv.diem_GPA = float(diem_GPA_moi)

                # Sửa thông tin riêng
                if isinstance(sv, SinhVienChatLuongCao):
                    diem_ta_moi = input(f"Điểm tiếng Anh mới ({sv.diem_tieng_anh}): ").strip()
                    if diem_ta_moi:
                        sv.diem_tieng_anh = float(diem_ta_moi)

                print("Cập nhật thông tin thành công!")
            except ValueError:
                print("Lỗi: Điểm nhập vào phải là một con số.")
        else:
            print("LỖI: Không tìm thấy sinh viên có mã này.")

    def xem_danh_sach_sinh_vien(self):
        """Chức năng 4: Xem danh sách sinh viên."""
        print("\n--- Danh Sách Sinh Viên ---")
        if not self.danh_sach_sv:
            print("Hiện chưa có sinh viên nào trong danh sách.")
            return

        for sv in self.danh_sach_sv:
            sv.hien_thi_thong_tin()
            # --- ĐÃ XÓA dòng in điểm thưởng ---
            print("---------------------------")

    def danh_gia_gpa(self):
        """Chức năng 5: Đánh giá xếp loại GPA."""
        print("\n--- Đánh Giá Xếp Loại GPA ---")

        ma_sv = input("Nhập mã SV cần đánh giá: ").strip()
        sv = self.tim_sinh_vien_theo_ma(ma_sv)

        if sv:
            print(f"Sinh viên: {sv.ho_ten}")

            diem_GPA = sv.diem_GPA
            if diem_GPA is None:
                print("LỖI: Sinh viên này chưa có điểm GPA.")
                return

            xep_loai = ""
            if diem_GPA >= 3.6:
                xep_loai = "Xuất sắc"
            elif diem_GPA >= 3.2:
                xep_loai = "Giỏi"
            elif diem_GPA >= 2.5:
                xep_loai = "Khá"
            elif diem_GPA >= 2.0:
                xep_loai = "Trung bình"
            else:
                xep_loai = "Yếu"

            print(f"Điểm GPA của {sv.ho_ten}: {diem_GPA}")
            print(f"Đạt loại: {xep_loai}")
        else:
            print("LỖI: Không tìm thấy sinh viên có mã này.")

# --- HÀM MAIN ĐỂ CHẠY CHƯƠNG TRÌNH ---
def main():
    qlsv = QuanLySinhVien()

    # --- Cập nhật data demo để test ---
    sv1 = SinhVienDaiTra("3122411058", "Đinh Trung Hội", "10/10/2004", 85, 2.8)
    sv2 = SinhVienChatLuongCao("3122411073", "Huỳnh Phúc Hưng", "20/07/2004", 80, 2.6, 8.4)

    qlsv.danh_sach_sv.extend([sv1, sv2])

    while True:
        print("\n===== CHƯƠNG TRÌNH QUẢN LÝ SINH VIÊN =====")
        print("1. Thêm sinh viên")
        print("2. Xóa sinh viên")
        print("3. Sửa thông tin sinh viên")
        print("4. Xem danh sách sinh viên")
        print("5. Đánh giá xếp loại GPA")
        print("0. Thoát chương trình")

        lua_chon = input("Nhập lựa chọn của bạn: ").strip()

        if lua_chon == '1':
            qlsv.them_sinh_vien()
        elif lua_chon == '2':
            qlsv.xoa_sinh_vien()
        elif lua_chon == '3':
            qlsv.sua_thong_tin_sinh_vien()
        elif lua_chon == '4':
            qlsv.xem_danh_sach_sinh_vien()
        elif lua_chon == '5':
            qlsv.danh_gia_gpa()
        elif lua_chon == '0':
            print("Đã thoát chương trình.")
            break
        else:
            print("Lựa chọn không hợp lệ. Vui lòng nhập lại.")

# --- Thực thi chương trình ---
if __name__ == "__main__":
    main()
```
