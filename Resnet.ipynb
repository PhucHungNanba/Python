```python
from tensorflow.keras.applications.resnet50 import ResNet50, decode_predictions
from tensorflow.keras.preprocessing import image
import numpy as np

# 1. Tải model đã được train sẵn trên ImageNet
model = ResNet50(weights='imagenet')

# 2. Đưa ảnh vào dự đoán (Giả sử có file 'test.jpg')
img = image.load_img('Hải cẩu.jpg', target_size=(224, 224))
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)

# 3. Ra kết quả và xử lý cho đẹp
preds = model.predict(x)
results = decode_predictions(preds, top=3)[0]

print("--- PHÂN TÍCH TỶ LỆ ---")
for i, (id, label, prob) in enumerate(results):
    # prob là xác suất (0.57), nhân 100 sẽ ra phần trăm (57.00%)
    print(f"{i+1}. {label}: {prob * 100:.2f}%")

# Lấy kết quả cao nhất (là cái đầu tiên trong list)
best_result = results[0]
print(f"\n=> KẾT QUẢ SAU CÙNG: Đây là con {best_result[1]} (Độ tin cậy: {best_result[2]*100:.2f}%)")
```
